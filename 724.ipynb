### 問題

実数$a$，$b$，$c$ が $a^2+b^2+c^2=1$ を満たして変化するとき，空間内の動点P $(2a+b,\,2c+b,\,a+b+c)$ が動くことができる部分の面積を求めよ。

### 解答ではないのですが．．．

1次変換$f : \begin{pmatrix}
        2 & 1 & 0  \\
        0 & 1 & 2  \\
        1 & 1 & 1  \\
    \end{pmatrix}$による単位球面の像を描いてみます。

$$\begin{cases}
    x&=2\cos v\cos u+\cos v\sin u\\
    y&=\cos v\sin u+2\sin v\\
    z&=\cos v\cos u+\cos v\sin u+\sin v\\
    \end{cases}$$

```julia
using PlotlyJS
```

```julia
using PlotlyJS

dom = range(0, stop=2π, length=100)
u = dom' .* ones(100)
v = ones(100)' .* dom


x = @. (2 * cos(v)*cos(u)+cos(v)*sin(u))
y = @. (cos(v)*sin(u)+2*sin(v))
z = @. (cos(v)*cos(u)+cos(v)*sin(u)+sin(v))

q = make_subplots(
    rows=1, cols=2,
    specs=[Spec(kind="scene") Spec(kind="scene")],
    subplot_titles=["Color corresponds to z"  "Color corresponds to distance to origin"]
)
add_trace!(q, surface(x=x, y=y, z=z, colorbar_x=-0.07), row=1, col=1)
add_trace!(q, surface(x=x, y=y, z=z, surfacecolor=@. x^2 + y^2 + z^2), row=1, col=2)
relayout!(q, title_text="楕円")
q
```
